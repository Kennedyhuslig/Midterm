```python
import matplotlib.pyplot as plt
import numpy as np
import statistics

data=[304,220,255,276,302,405,210,288,297,310,420,300,270,306,209,266,269,305,270,306,335,278,299,330,400,357,306,279,305,306,322]

plt.hist(data,bins=range(125,500,60), align='left')
plt.xlabel('Screen Time (minutes)')
plt.ylabel('Frequency')
plt.title('Histogram of Screen Time')
plt.show()

mean=np.mean(data)
median=np.median(data)
mode=statistics.mode(data)
standard_dev=np.std(data)

plt.hist(data,bins=range(125,500,60),align='left')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.axvline(mean, color='r', linewidth=1)
plt.axvline(median, color='g',linewidth=1)
plt.axvline(mode, color='b', linewidth=1)
plt.axvline(standard_dev, color='y',linewidth=1)
plt.show()
```
